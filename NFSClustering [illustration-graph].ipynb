```python
threshold = []
activation = ["d11","d12","dbs","dbd","d71","d72"]
for key in activation:
    memat = mmat(data,key)
    comat = cmat(memat)
    eqvmat = emat(memat)
    threshold.append(lbd(eqvmat))
print(np.array(threshold))
```

```python
x = [1,2,3,4,5]
l = [1,2,7]
for i in range(6):
    if i < 3:
        plt.plot(x,threshold[i],label = "D{} to S'{}".format(l[i],1))
    else:
        plt.plot(x,threshold[i],label = "D{} to S'{}".format(l[i-3],2))
plt.xticks(x)
plt.legend(loc='lower right')
plt.xlabel("Number of clusters")
plt.ylabel("Threshold of confidence levels")
plt.title("The correlation between the number of clusters and the threshold of confidence levels")
plt.show()
```
